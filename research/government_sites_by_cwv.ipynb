# Ranking per origin

Here, we're examining individual origins based on the Chrome User Experience (CrUX) Report. The scope is limited to the origins contained in the site scanning report from the export the week of July 22, 2024 who are also present in the CrUX data set and limiting it to mobile data.

```python
"""
Loads result to dataframe and analyzes the results
"""

import os
import pandas as pd
from format_urls.load_results_to_dataframe import load_results_to_dataframe


def filter_results(**kwargs):
    """load results to dataframe"""
    df = load_results_to_dataframe(
        file_name=kwargs["file_name"], notebook_dir=kwargs["notebook_dir"]
    )

    if kwargs["filter_column"] is not None:
        df = df[df[kwargs["filter_column"]] == kwargs["form_factor"]].sort_values(
            by=kwargs["sort_by"], ascending=False
        )
    df.reset_index()
    df.set_index(kwargs["index_on"], inplace=True)
    return df
```

```python
cwv = filter_results(
    file_name="data/cwv-by-origin-and-device-202407.csv",
    index_on="origin",
    sort_by="good_lcp",
    filter_column="device",
    form_factor="phone",
    notebook_dir=os.getcwd(),
)

cwv.head(5000)
```

```python
lcp = (
    cwv[["good_lcp", "needs_improvement_lcp", "poor_lcp", "p75_lcp"]]
    .dropna(subset=["good_lcp"])
    .sort_values(by="good_lcp", ascending=True)
)
lcp.head(10000)
```

```python
cls = (
    cwv[["good_cls", "needs_improvement_cls", "poor_cls", "p75_cls"]]
    .dropna(subset=["good_cls"])
    .sort_values(by="good_cls", ascending=True)
)
cls.head(10000)
```

```python
inp = (
    cwv[["good_inp", "needs_improvement_inp", "poor_inp", "p75_inp"]]
    .dropna(subset=["good_inp"])
    .sort_values(by="good_inp", ascending=True)
)
inp.head(10000)
```

```python
notebook_dir = os.getcwd()

site_scan = load_results_to_dataframe(
    notebook_dir=notebook_dir,
    file_name="data/site-scanning-weekly-snapshot-20240722.csv",
)
site_scan["final_url_domain_https"] = "https://" + site_scan["final_url_website"]
site_scan = site_scan[
    ["final_url_domain_https", "target_url_agency_owner"]
].drop_duplicates(subset="final_url_domain_https")
```

## Getting agency-level detail

One of the things we during research interviews was that stakeholder reporting is an important component of web performance optimization efforts and is something that often gets overlooked. Here, we are compiling the data by agency to get a picture of performance. Once they are grouped by agency, that helps narrow down which ones we might want to examine in greater detail.

```python
cwv_with_agency = pd.merge(
    cwv, site_scan, left_on="origin", right_on="final_url_domain_https", how="inner"
)

summary = (
    cwv_with_agency.groupby("target_url_agency_owner")
    .agg(
        {
            "p75_inp": "median",
            "p75_lcp": "median",
            "p75_cls": "median",
            "target_url_agency_owner": "count",
        }
    )
    .rename(columns={"target_url_agency_owner": "count"})
    .reset_index()
)

summary = summary.rename(
    columns={
        "p75_inp": "median_p75_inp",
        "p75_lcp": "median_p75_lcp",
        "p75_cls": "median_p75_cls",
    }
)

summary.head(5000)
```
